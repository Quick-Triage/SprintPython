```python
from random import randrange
 
def pergunta_febre():
    tem_febre = str(input("O paciente apresenta sintomas de febre? ")).lower()
 
    temperatura = None
    if "sim" in tem_febre:
        temperatura = str(input("Qual a temperatura mais alta que o paciente apresentou? ")).lower()
 
    pressao = str(input("O paciente possui problemas com pressão? "))
 
    problema_pressao = None
 
    if "sim" in pressao:
        problema_pressao = str(input("Problemas com pressão alta ou baixa? "))
 
    return tem_febre, pressao, problema_pressao, temperatura
 
def pergunta_sintomas():
    dores = str(input("O paciente possui dor em alguma parte do corpo? ")).lower()
 
    local_dor = None
    motivo_dor = None
 
    if "sim" in dores:
        local_dor = str(input("Onde o paciente sente dor? "))
        motivo_dor = str(input("O paciente sabe o motivo da dor, se sim qual? "))
 
    vomito = str(input("O paciente apresenta sintomas de vômito? "))
   
    return dores, local_dor, motivo_dor, vomito
 
def triagem():
    nome = str(input("Digite seu nome ou o nome completo do paciente: "))
    cpf = str(input("Digite seu CPF ou o CPF do paciente: "))
    if len(cpf) != 11:
        raise ValueError("CPF inválido. O CPF deve ter 11 caracteres.")
    idade = int(input("Digite sua idade ou a idade do paciente: "))
 
    nome_resp = None
 
    if idade < 18:
        nome_resp = str(input("Digite o nome do responsável: "))
        cpf_resp = str(input("Digite o CPF do responsável: "))
        if len(cpf_resp) != 11:
            raise ValueError("CPF inválido. O CPF deve ter 11 caracteres.")
        idade_resp = int(input("Digite a idade do responsável: "))
 
    febre, pressao, problema_pressao, temperatura = pergunta_febre()
 
    dor, local_dor, motivo_dor, vomito = pergunta_sintomas()
 
    data_sintomas = str(input("Quando os sintomas começaram? "))
 
    ficha_medica = {
        "Nome": nome,
        "CPF": cpf,
        "Idade": idade,
        "Pressão": f"{pressao}, {problema_pressao if "sim" in pressao else ''}",
        "Vômito": vomito,
        "Febre": f"{febre}, {temperatura if 'sim' in febre else ''}",
        "Dores no corpo": f"{dor}, {local_dor if 'sim' in dor else ''}",
        "Quando começaram os sintomas": data_sintomas,
    }
 
    if "sim" in dor:
        ficha_medica['Sabe o motivo da dor'] = motivo_dor
 
    return ficha_medica, nome_resp
 
def exibir_ficha(ficha, nome_resp):
    print("\n------- HOSPITAL SABARA -------\n")
    numero_chamada = randrange(0, 200)
    print(f"\n---- Numero de chamada: {numero_chamada} ----\n")
   
    if nome_resp:
        print(f"Nome do responsável: {nome_resp}")
 
    for chave, valor in ficha.items():
        print(f"{chave}: {valor}")
 
    print("-" * 40)
 
ficha, nome_resp = triagem()
exibir_ficha(ficha, nome_resp)
```
